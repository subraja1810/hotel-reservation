```python
# Ask the customer whether they've come for booking or to vacate.
# If for booking, Display list of rooms with prices and availability. 
# Ask them to choose, set availability to false, ask no. of days of stay, and ask them to pay.
# If vacate, ask details of their room and set availability to true.
```

```python
from tabulate import tabulate
```

```python
single_avail = 10
double_avail = 8
queen_avail = 6
mini_avail = 4
penthouse_avail = 2
```

```python
def display_rooms(): # display to customer, details of rooms and availability
    room_details = [['TYPE','PRICE in Rs.','NO. OF AVAILABLE ROOMS'],['Single BedRoom',1000,single_avail],['Double BedRoom',2000,double_avail],['Queen',5000,queen_avail],['Mini Suite',8000,mini_avail],['Penthouse Suite',20000,penthouse_avail]]
    print(tabulate(room_details, headers='firstrow', tablefmt='fancy_grid'))
```

```python
#display_rooms()
```

```python
def price_calc(): # calculate bill for customer
   
    valid_choice = True
    
    while valid_choice:
        
        room_choice = input("Which type of room do you prefer? ")
        no_of_rooms = int(input("Enter the number of rooms: "))
        no_of_days = int(input("Enter the number of days you wish to stay: "))
        
        if room_choice.lower() == 'single bedroom':
            total_price = 1000 * no_of_days * no_of_rooms
            valid_choice = False
            
        elif room_choice.lower() == 'double bedroom':
            total_price = 2000 * no_of_days * no_of_rooms
            valid_choice = False
            
        elif room_choice.lower() == 'queen':
            total_price = 5000 * no_of_days * no_of_rooms
            valid_choice = False
            
        elif room_choice.lower() == 'mini suite':
            total_price = 8000 * no_of_days * no_of_rooms
            valid_choice = False
            
        elif room_choice.lower() == 'penthouse suite':
            total_price = 20000 * no_of_days * no_of_rooms
            valid_choice = False
            
        else:
            valid_choice = True
               
    return (room_choice,no_of_rooms,no_of_days,total_price)
```

```python
def display_bill(room_choice,no_of_rooms,no_of_days,total_price): 
    booking_details = [['TYPE','NO. OF ROOMS','NO. OF DAYS','TOTAL PRICE'],[room_choice,no_of_rooms,no_of_days,total_price]]
    print(tabulate(booking_details, headers='firstrow', tablefmt='fancy_grid'))
    print("Please pay the bill! Have a nice stay.")
```

```python
again = True

while again:

    print("Welcome to Hotel W!")

    customer_choice = input("Do you want to reserve rooms or vacate? ")

    if customer_choice == 'reserve':

        display_rooms()

        proceed = input("Do you wish to continue booking? (y or n): ")

        if proceed.lower() == 'y':
            reserve = True
        else:
            reserve = False

        while reserve:

            room_choice,no_of_rooms,no_of_days,total_price = price_calc()
            
            available = True

            if room_choice.lower() == 'single bedroom' and single_avail >= no_of_rooms:
                single_avail -= no_of_rooms
                
            elif room_choice.lower() == 'double bedroom' and double_avail >= no_of_rooms:
                double_avail -= no_of_rooms

            elif room_choice.lower() == 'queen' and queen_avail >= no_of_rooms:
                queen_avail -= no_of_rooms

            elif room_choice.lower() == 'mini suite' and mini_avail >= no_of_rooms:
                mini_avail -= no_of_rooms

            elif room_choice.lower() == 'penthouse suite' and penthouse_avail >= no_of_rooms:
                penthouse_avail -= no_of_rooms
                    
            else:
                available = False
                print("Sorry! Not enough rooms available")
   

            if available == True:
                
                display_bill(room_choice,no_of_rooms,no_of_days,total_price)
                break
                
            break   

    elif customer_choice == 'vacate':

        vacate = True

        while vacate:

            details = input("Enter room type: ")
            no_of_room = int(input("Enter the no. of rooms: "))

            if details.lower() == 'single bedroom' and (single_avail + no_of_room) <= 10:
                single_avail += no_of_room
                vacate = False

            elif details.lower() == 'double bedroom' and (double_avail + no_of_room) <= 8:
                double_avail += no_of_room
                vacate = False

            elif details.lower() == 'queen' and (queen_avail + no_of_room) <= 6:
                queen_avail += no_of_room
                vacate = False

            elif details.lower() == 'mini suite' and (mini_avail + no_of_room) <= 4:
                mini_avail += no_of_room
                vacate = False

            elif details.lower() == 'penthouse suite' and (penthouse_avail + no_of_room) <= 2:
                penthouse_avail += no_of_room
                vacate = False

            else:
                vacate = True   

        print("Thanks! Hope you enjoyed your stay here.")
        
    else:
        print("Thanks for visiting!")
        break
```
